```python
string = '12:45:54AM'


def timeConversion(s):
    strings = s.split(':')
    hour = int(strings[0])
    mins = (strings[1])
    secs = (strings[2][:2])
    
    if strings[2][2:4] =='PM'and hour != 12:
        hour = hour + 12
        print(f"{hour}:{mins}:{secs}")
    if strings[2][2:4] =='PM'and hour == 12:
        print(f"{strings[0]}:{mins}:{secs}")
    if strings[2][2:4] =='AM' and hour !=12:
        print(f"{strings[0]}:{mins}:{secs}")
    if strings[2][2:4] =='AM' and hour ==12:
        print(f"00:{mins}:{secs}")
        
if __name__ == '__main__':
    timeConversion(string)
```
