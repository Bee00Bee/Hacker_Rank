```python
def gradingStudents(grades):
    results = []
    for grade in grades:
        divisions = divmod(grade, 5)
        multiple_five = (divisions[0]+1)*5
        if grade < 38:
            results.append(grade)
        elif multiple_five - grade < 3 and multiple_five-grade >0:
            results.append(multiple_five)
        elif multiple_five-grade >=3 or multiple_five-grade <0:
            results.append(grade)
    return results
```

```python
grades = [14, 30, 24, 39, 70]
gradingStudents(grades)
```
