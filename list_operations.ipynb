```python
def lst_process(order_lst):
    lst_out = []

    for lst in order_lst:
        splitted_lst = lst.split()
  
        if splitted_lst[0] == 'append':
            #print(splitted_lst[1])
            lst_out.append(int(splitted_lst[1]))
        elif splitted_lst[0] == 'insert':
            lst_out.insert(int(splitted_lst[1]), int(splitted_lst[2]))
        elif splitted_lst[0] == 'remove':
            lst_out.remove(int(splitted_lst[1]))
        elif splitted_lst[0] == 'print':
            print(lst_out)
        elif splitted_lst[0] == 'sort':
            lst_out.sort()
        elif splitted_lst[0] == 'pop':
            lst_out.pop()
        elif splitted_lst[0] == 'reverse':
            lst_out.reverse()


if __name__ == '__main__':
    N = int(input())
    
    lst = []
    while N != 0:
        lst.append(input())
        N -= 1

    lst_process(lst)
```
