```python
def palindromic(n):
    lst = []
    string = '' 

    while n!=0:
        m = n

        string =''
        for i in range(1, m+1):
            if i <n:
                string += str(i)

        while m != 0:
            string += str(m)
            m -= 1

        lst.append(int(string))

        n -= 1

    [print(ele) for ele in lst[::-1]]
    
if __name__=='__main__':
    n = int(input())
    palindromic(n)
```
