```python
def dic_process(lst):
    
    dic = dict()
    names = []
    
    for ele in lst:
        eles = ele.split(' ')
        names.append(' '.join(eles[:-1]))
        
    names2 = []
    
    for name in names:
        if name not in names2:
            names2.append(name)
    
    for ele in names2:
        dic[ele] = 0
        
    for ele in lst:
        eles = ele.split(' ')
        name = ' '.join(eles[:-1])
        dic[name] += int(eles[-1])
    
    for key, value in dic.items():
        print(key, value)
        
if __name__ == '__main__':
    
    N = int(input())
    
    lst = []
    while N !=0:
        lst.append(input())
        N -= 1
        
    dic_process(lst)
```
